```python
# Loading the dataset using pandas

import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head() #top 5 rows so we can understand the data
```

```python
df.info() #to get structure of data
```

```python
# Summary statistics using .describe() by default it gives summary statistics of numerical columns only and if want to get for other columns too need to write include='all'
df.describe(include='all')
```

```python
#Data Preprocessing
```

```python
# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category We have not chosen mean as it can get affected by outliers and that could skew the distribution of data and median is robust to outliers.We donot use the same median for all the mising values as that can cause bias in our dataset coz the dataset has differenst categories so we will find the median review rating within each category and replace it with missing values

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation
#snake case means lower case with _

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Feature Engineering
```

```python
# create a new column age_group ; qcut splits the ages into 4 equal sized grps based on data distribution
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days converting textual features to numbers to make analysis easier

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
